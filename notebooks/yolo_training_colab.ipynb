**STEP 5:**

___________________________________________________________

````python
from ultralytics import YOLO

model = YOLO("yolov8n.pt")

model.train(
    data="/content/dataset/data.yaml",
    epochs=15,
    imgsz=512,
    batch=32,
    device=0,   # GPU
    workers=2,
    project="/content/runs",
    name="ppe_yolo_colab",
    seed=42
)
````
